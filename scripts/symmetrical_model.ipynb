```python
# Import libraries
import moments, moments.LD
import pandas as pd, numpy as np
from collections import defaultdict
import demes, demesdraw, matplotlib.pylab as plt

import model, slice, get_LD, plot_LD
```

```python
# Create symmetric model
N = 30e3
time_period = 0.1*N
cake = model.cake_model(Ns=[N,N,N],splits = [3,3],proportions=[[1/3,1/3,1/3],[1/3,1/3,1/3]],time_period_merge=[time_period,time_period,time_period],time_period_splits=[time_period,time_period],yaml_filename="../yaml_files/symmetrical_model.yaml",plot_filename="../Figures/symmetric_cake/symmetric_cake")
```

```python
# Slice
times,times_dic = slice.sample_timepoints(cake)
sliced_dict = slice.slice_timepoints(cake,times,yaml_filename="../yaml_files/symmetrical_model")
```

```python
# Compute LD
LD_dict = get_LD.get_LD_from_sliced_demes(sliced_dict)
```

```python
# Plot LD
rhos = np.logspace(-2, 2, 21)
plot_LD.plot_LD(LD_dict,rhos,times_dic,LD_dict['Ancestral'][0],plot_file="../Figures/symmetric_cake/symmetrical_model.svg")
```
